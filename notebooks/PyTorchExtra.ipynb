```python
import sys
import os
import torch
import torch.nn as nn 
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from torch.utils.data import random_split

sys.path.append(os.path.abspath('../src'))

from DataLoader import DataLoader as DataLoader__
from DataSplitter import DataSplitter
```

```python
class MNIST(Dataset):
    def __init__(self, path, label=None):
        df = pd.read_csv(path)
        self.label = label

        if self.label is not None:
            self.X = torch.tensor(
                df.drop(columns=[self.label]).values,
                dtype=torch.float32
            )
            self.y = torch.tensor(
                df[self.label].values,
                dtype=torch.long
            )
        else:
            self.X = torch.tensor(
                df.values,
                dtype=torch.float32
            )

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        if self.label:
            return self.X[idx], self.y[idx]

        return self.X[idx]
```

```python
path_train, path_test = "../data/train.csv", "../data/test.csv"
dataset_train = MNIST(path=path_train, label='label')
dataset_test = MNIST(path=path_test)
print(len(dataset_train), len(dataset_test))

## train test split
train_dataset, eval_dataset = random_split(dataset_train, [0.8, 0.2])
print(len(train_dataset), len(eval_dataset), len(eval_dataset)/(len(eval_dataset) + len(train_dataset)))
X_train = train_dataset.dataset.X[train_dataset.indices]
y_train = train_dataset.dataset.y[train_dataset.indices]
```

```python
## DataLoader
batch_size = 200
data_loader = DataLoader(
    dataset=dataset_train,
    batch_size=batch_size,
    shuffle=True,
    num_workers=0,       # Set higher (e.g., 2 or 4) for faster multi-process I/O
    pin_memory=False     # Set True if training directly on CUDA GPUs
)
for batch_idx, (features, labels) in enumerate(data_loader):
    if(batch_idx % 100 == 0):
        print(f"Batch Idx: {batch_idx}, Shape features: {features.shape}, labels: {labels}")
```
